```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('data_domestic.csv')
```

```python
df.info()
```

```python
df["localtime"] = pd.to_datetime(df["localtime"])
```

```python
df.groupby(pd.Grouper(key='localtime', freq='30min')).count()
```

```python
df
```

```python
df_temp = df.drop_duplicates(subset ="ClientMacAddr", 
                     keep = 'first', inplace = False)
```

```python
df_temp
```

```python
df_count = df_temp.groupby(pd.Grouper(key='localtime', freq='30min')).count()
```

```python
df_discount = df_count.drop(['Unnamed: 0','Building','Level','lat','lng','datetime'],axis=1)
```

```python
df_discount.columns
```

```python
df_discount= df_discount.rename(columns = {'ClientMacAddr' : 'Distinct_Macid'})
```
