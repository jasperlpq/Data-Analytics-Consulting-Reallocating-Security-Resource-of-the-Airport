```python
import json

yourData = []
with open('./teca_adm_and_exp/teca_adm_exp_000000000000') as inputData:
    for line in inputData:
        try:
            yourData.append(json.loads(line.rstrip('\n')))
        except ValueError:
            print("Skipping invalid line {0}").format(repr(line))

print (yourData)
```
